## EDA

#### As our columns in string format (Categorical) we have to covert them into numerical value to train a model on the data.

```python
df['Journey_day']=pd.to_datetime(df.Date_of_Journey,format="%d/%m/%Y").dt.day   # getting day of journey
```

```python
df['Journey_month']=pd.to_datetime(df.Date_of_Journey,format="%d/%m/%Y").dt.month    # getting month of journey
```

```python
df.head()
```

#### Journey year can also be one feature but this data is of only 2019

```python
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df["Dep_hour"]=pd.to_datetime(df['Dep_Time']).dt.hour
df["Dep_min"]=pd.to_datetime(df['Dep_Time']).dt.minute
df.drop('Dep_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_hour']=pd.to_datetime(df.Arrival_Time).dt.hour
df['Arrival_min']=pd.to_datetime(df.Arrival_Time).dt.minute
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df.head()
```

```python
# Converting duration into minutes
durat=list(df.Duration)

for i in range(len(durat)):
    if "h" in durat[i]:
        li=durat[i].split("h")
        if li[1]=="":
            durat[i]=60*int(li[0])
        else:
            li[1]=li[1].replace("m","")
            durat[i]=60*int(li[0])+int(li[1])
    else:
        durat[i]=durat[i].replace("m","")
        durat[i]=int(durat[i])
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
df['Duration']=durat
```

```python
df.head()
```

#### From graph we can see that Jet Airways Business have the highest Price.

```python
# One hot encoding of airline feature
Airline = df.Airline
Airline = pd.get_dummies(Airline, drop_first= True)
Airline.head()
```

```python
sns.catplot(y = "Price", x = "Source", data = df.sort_values("Price", ascending = False), kind="boxen", height = 4, aspect = 3)
plt.show()
```

```python
#One hot encoding of source feature (Nominal Feature)
source=df.Source
source=pd.get_dummies(source,drop_first=True)
source.head() 
```

```python
df.Destination.value_counts()
```

```python
sns.catplot(y = "Price", x = "Destination", data = df.sort_values("Price", ascending = False), kind="boxen", height = 4, aspect = 3)
plt.show()
```

```python
Destination = df.Destination  # Destination : Nominal
Destination = pd.get_dummies(Destination, drop_first = True)
Destination.head()
```

```python
df.Route.value_counts()
```

#### Here route and number of stops are doing the same thing so we can drop route feature

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.Additional_Info.value_counts()
```

```python
sns.catplot(y = "Price", x = "Additional_Info", data = df.sort_values("Price", ascending = False), kind="boxen", height = 4, aspect = 3)
plt.show()
```

#### From the graph we can say that the price of business class is high 

```python
add_info=df.Additional_Info
add_info=pd.get_dummies(add_info,drop_first=True)
add_info.head()
```

```python
df.Total_Stops.value_counts()
```

#### We can do ordinal encoding of stops feature because price increase or decrease with increase or decrease in stops

```python
def label(stops):
    omap={'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4}
    return omap[stops]
```

```python
df.Total_Stops=df.Total_Stops.apply(label)
```

# Test Set 

```python
test_data=pd.read_excel("Test_set.xlsx")
```

```python
test_data.head()
```

```python
# Preprocessing

print("Test data Info")
print("-"*75)
print(test_data.info())

print()
print()

print("Null values :")
print("-"*75)
test_data.dropna(inplace = True)
print(test_data.isnull().sum())

# EDA

# Date_of_Journey
test_data["Journey_day"] = pd.to_datetime(test_data.Date_of_Journey, format="%d/%m/%Y").dt.day
test_data["Journey_month"] = pd.to_datetime(test_data["Date_of_Journey"], format = "%d/%m/%Y").dt.month
test_data.drop(["Date_of_Journey"], axis = 1, inplace = True)

# Dep_Time
test_data["Dep_hour"] = pd.to_datetime(test_data["Dep_Time"]).dt.hour
test_data["Dep_min"] = pd.to_datetime(test_data["Dep_Time"]).dt.minute
test_data.drop(["Dep_Time"], axis = 1, inplace = True)

# Arrival_Time
test_data["Arrival_hour"] = pd.to_datetime(test_data.Arrival_Time).dt.hour
test_data["Arrival_min"] = pd.to_datetime(test_data.Arrival_Time).dt.minute
test_data.drop(["Arrival_Time"], axis = 1, inplace = True)

# Duration
durat=list(test_data.Duration)

for i in range(len(durat)):
    if "h" in durat[i]:
        li=durat[i].split("h")
        if li[1]=="":
            durat[i]=60*int(li[0])
        else:
            li[1]=li[1].replace("m","")
            durat[i]=60*int(li[0])+int(li[1])
    else:
        durat[i]=durat[i].replace("m","")
        durat[i]=int(durat[i])
test_data.Duration=durat

# Categorical data

print("Airline")
print("-"*75)
print(test_data["Airline"].value_counts())
Airline = pd.get_dummies(test_data["Airline"], drop_first= True)

print()

print("Source")
print("-"*75)
print(test_data["Source"].value_counts())
Source = pd.get_dummies(test_data["Source"], drop_first= True)

print()
add_info=test_data.Additional_Info
add_info=pd.get_dummies(add_info,drop_first=True)

print("Destination")
print("-"*75)
print(test_data["Destination"].value_counts())
Destination = pd.get_dummies(test_data["Destination"], drop_first = True)

# Additional_Info contains almost 80% no_info
# Route and Total_Stops are related to each other
test_data.drop(["Route"], axis = 1, inplace = True)

# Replacing Total_Stops
test_data.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)

# Concatenate dataframe --> test_data + Airline + Source + Destination
data_test = pd.concat([test_data,Source,Destination,Airline,add_info], axis = 1)

data_test.drop(["Airline", "Source", "Destination"], axis = 1, inplace = True)

print()
print()

print("Shape of test data : ", data_test.shape)
```
